```python
# Import Libraries
import pandas as pd
from dateutil import parser
import matplotlib.pyplot as plt
```

```python
path = "REPLACE WITH FILE PATH"
```

```python
# Import Data

scrape_data = pd.read_csv(path + "scrape_data.csv")
```

```python
scrape_data.head()
```

```python
scrape_data['published_date'] = scrape_data['published_date'].apply(lambda x: parser.parse(x))
```

```python
authors_count = scrape_data.groupby(["author"]).size().reset_index(name='counts').sort_values(["counts"], ascending=False)
```

```python
authors_count.head(10)
```

```python
# Need to expand dates - the chart only shows occurences but not a complete timeline
scrape_data.groupby(scrape_data['published_date'].dt.date).size().plot(figsize=(18, 10))
```
