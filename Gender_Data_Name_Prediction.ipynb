```python
# Import Libraries
import pandas as pd
from dateutil import parser
import matplotlib.pyplot as plt
```

```python
path = "REPLACE WITH FILE PATH"
```

```python
# Import Data

scrape_data = pd.read_csv(path + "scrape_data.csv")
```

```python
scrape_data.head()
```

```python
scrape_data['published_date'] = scrape_data['published_date'].apply(lambda x: parser.parse(x))
```

```python
import nltk
import random

nltk.download('names')

def extract_features(name):
    name = name.lower()
    return {
        'last_char': name[-1],
        'last_two': name[-2:],
        'last_three': name[-3:],
        'first': name[0],
        'first2': name[:1]
    }

f_names = nltk.corpus.names.words('female.txt')
m_names = nltk.corpus.names.words('male.txt')

all_names = [(i, 'm') for i in m_names] + [(i, 'f') for i in f_names]
random.shuffle(all_names)

test_set = all_names[3200:]
train_set= all_names[:3200]

test_set_feat = [(extract_features(n), g) for n, g in test_set]
train_set_feat= [(extract_features(n), g) for n, g in train_set]

classifier = nltk.NaiveBayesClassifier.train(train_set_feat)

print(nltk.classify.accuracy(classifier, test_set_feat))
```

```python
classifier.classify(extract_features('Tiffany'))
```

```python
authors_count = scrape_data.groupby(["author"]).size().reset_index(name='counts').sort_values(["counts"], ascending=False)
authors_count
```

```python
authors_count['first_name'] = authors_count['author'].apply(lambda x: x.split()[0])
authors_count
```

```python
authors_count['name_prediction'] = authors_count['first_name'].apply(lambda x: classifier.classify(extract_features(x)))
```
